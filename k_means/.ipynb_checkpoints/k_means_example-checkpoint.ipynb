# K_means clustering

## import libraries

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## Creating Data

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=300, n_features=3, centers=4, cluster_std=1.4, random_state=101)
```

## Visualizing Data

```python
plt.scatter(data[0][:,0], data[0][:,1], c=data[1])
```

## Creating the clusters

### With implemented K_means

```python
import k_means_cluster
groups, centers=k_means_cluster.k_means_iter(data[0],4,100)
```

```python
groups
```

```python
centers
```

### With Scikit Learn

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

### Comparing results

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True,figsize=(15,7))
ax1.set_title('implemented K_means')
ax1.scatter(data[0][:,0],data[0][:,1],c=groups,cmap='rainbow')
ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
ax3.set_title('K Means (sklearn)')
ax3.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')
```
